### What are the top rated movies according to rotten tomatoes ?

```python
data.rename(columns = {'Rotten Tomatoes' : 'RottenTomatoes'}, inplace =True)
```

```python
data.sort_values(by="RottenTomatoes", ascending=False)
```

### Graph no of movies categorized by age restrictions before 2000 and after 2000 ?

```python
Yeargroups  = data.Year.unique()
```

```python
Yeargroups
```

```python
Yeargroups.min()
```

```python
Before2000 = data[(data.Year <=2000)].head()
```

```python
Before2000
```

```python
After2000 = data[(data.Year > 2000)]
```

```python
After2000.count()
```

```python
MoviesBefore2000 = Before2000.groupby('Age').Title.count()
```

```python
MoviesBefore2000
```

```python
MoviesAfter2000 = After2000.groupby("Age").Age.count()
```

```python
MoviesAfter2000
```

```python
plt.plot(MoviesBefore2000,'r--')
plt.plot(MoviesAfter2000, 'g')
plt.xlabel(" Age categories ")
plt.ylabel(" no of movies ")
plt.show()
```
